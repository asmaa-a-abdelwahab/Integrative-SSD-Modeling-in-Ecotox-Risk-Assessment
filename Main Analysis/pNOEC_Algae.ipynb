```python
import pandas as pd
df=pd.read_csv ('finalNOEC.csv')
# Drop rows with NaN in the "palgae" column
df = df.dropna(subset=['pNOEC_Algae'])
df
```

```python
df_excluded = df.drop(columns=['BCUT2D_MWLOW', 'BCUT2D_MWUP']) #descriptors range excluded float does not work

# Define your selection: column 13 (index 12) and from column 23 onward
selected_columns = [df_excluded.columns[12]] + list(df_excluded.columns[22:])

# Select these columns
x = df_excluded[selected_columns]  # Only descriptors
x
y=df['pNOEC_Algae'] #endpoint
y.to_csv('YNOEC.csv', index=False)
y
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.feature_selection import SelectKBest, f_regression, mutual_info_regression, RFECV, SelectFromModel
from sklearn.decomposition import PCA
from tqdm import tqdm

# Load data

y = pd.read_csv('YNOEC.csv')

# Ensure y is 1D by converting it to a numpy array and flattening it
y = y.values.ravel()
print("Data loaded and y reshaped.")

# Step 1: Data Preprocessing
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print("Data split into training and testing sets.")

# Handle missing values with mean imputation
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)
print("Missing values imputed.")

# Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.transform(X_test_imputed)
print("Data scaled.")

# Feature Selection Techniques
feature_sets = {}

# Univariate Feature Selection (ANOVA)
selector_anova = SelectKBest(score_func=f_regression, k=10)
feature_sets['ANOVA'] = (selector_anova.fit_transform(X_train_scaled, y_train), selector_anova.transform(X_test_scaled))
print("Univariate feature selection using ANOVA completed.")

# Univariate Feature Selection (Mutual Information)
selector_mi = SelectKBest(score_func=mutual_info_regression, k=10)
feature_sets['Mutual Information'] = (selector_mi.fit_transform(X_train_scaled, y_train), selector_mi.transform(X_test_scaled))
print("Univariate feature selection using Mutual Information completed.")

# Lasso Regularization with increased max_iter
lasso = Lasso(alpha=0.01, max_iter=5000)  # Increase max_iter to ensure convergence
lasso.fit(X_train_scaled, y_train)
model = SelectFromModel(lasso, prefit=True)
feature_sets['Lasso'] = (model.transform(X_train_scaled), model.transform(X_test_scaled))
print("Feature selection using Lasso Regularization completed.")

# Feature Importance from Tree-based Models (Random Forest)
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train_scaled, y_train)
model = SelectFromModel(rf, prefit=True)
feature_sets['Random Forest'] = (model.transform(X_train_scaled), model.transform(X_test_scaled))
print("Feature selection using Random Forest feature importance completed.")

# Principal Component Analysis (PCA)
pca = PCA(n_components=10)
feature_sets['PCA'] = (pca.fit_transform(X_train_scaled), pca.transform(X_test_scaled))
print("PCA completed.")

Model Training and Evaluation
# Dictionary to store model scores for different feature selection techniques
model_scores = {}

# List of models to evaluate
models = {
    "Linear Regression": LinearRegression(),
    "Ridge": Ridge(alpha=1.0),
    "Lasso": Lasso(alpha=0.01, max_iter=5000),  # Also set max_iter for Lasso here
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42),
    "Support Vector Regressor": SVR()
}

# Evaluate each model on the different feature sets
for feature_set_name, (X_train_fs, X_test_fs) in feature_sets.items():
    print(f"\nEvaluating models on feature set: {feature_set_name}")
    model_scores[feature_set_name] = {}
    
    for model_name, model in models.items():
        model.fit(X_train_fs, y_train)
        train_score = model.score(X_train_fs, y_train)
        test_score = model.score(X_test_fs, y_test)
        model_scores[feature_set_name][model_name] = {"Train R2": train_score, "Test R2": test_score}
        
        # Print training and test scores for each model
        print(f"{model_name} - Train R²: {train_score:.4f}, Test R²: {test_score:.4f}")

# Step 8: Hyperparameter Tuning for Best Models (Random Forest and SVR using original feature set)
print("\nStarting hyperparameter tuning...")

# Random Forest Tuning
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_search_rf = GridSearchCV(RandomForestRegressor(random_state=42), param_grid_rf, cv=5, scoring='r2', n_jobs=-1)
grid_search_rf.fit(X_train_scaled, y_train)

# Best Random Forest Model and Scores
best_rf_model = grid_search_rf.best_estimator_
best_rf_params = grid_search_rf.best_params_
best_rf_train_score = best_rf_model.score(X_train_scaled, y_train)
best_rf_test_score = best_rf_model.score(X_test_scaled, y_test)
print("\nBest Random Forest Parameters:", best_rf_params)
print("Random Forest - Train R²:", best_rf_train_score, ", Test R²:", best_rf_test_score)

# Support Vector Regressor (SVR) Tuning
param_grid_svr = {
    'C': [0.1, 1, 10],
    'epsilon': [0.01, 0.1, 1],
    'kernel': ['linear', 'rbf']
}
grid_search_svr = GridSearchCV(SVR(), param_grid_svr, cv=5, scoring='r2', n_jobs=-1)
grid_search_svr.fit(X_train_scaled, y_train)

# Best SVR Model and Scores
best_svr_model = grid_search_svr.best_estimator_
best_svr_params = grid_search_svr.best_params_
best_svr_train_score = best_svr_model.score(X_train_scaled, y_train)
best_svr_test_score = best_svr_model.score(X_test_scaled, y_test)
print("\nBest SVR Parameters:", best_svr_params)
print("SVR - Train R²:", best_svr_train_score, ", Test R²:", best_svr_test_score)

# Final Output
print("\nModel Comparison After Tuning:")
print("Random Forest - Train R²:", best_rf_train_score, ", Test R²:", best_rf_test_score)
print("SVR - Train R²:", best_svr_train_score, ", Test R²:", best_svr_test_score)
```
