# Merging descrip, physchem, fingerprint_NOEC

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
df_phys=pd.read_csv('phys_chem_propertiesNOEC.csv')
df_phys = df_phys.drop_duplicates()
df_des=pd.read_csv('MaindatasetNOECdescriptors.csv')
df_des= df_des.drop_duplicates()
df_Afing= pd.read_csv('/datasetNOEC/AtomPair_fingerprints.csv')
df_Afing = df_Afing.drop_duplicates()
df_Mfing= pd.read_csv('/datasetNOEC/MACCS_fingerprints.csv')
df_Mfing = df_Mfing.drop_duplicates()
df_MOfing= pd.read_csv('/datasetNOEC/Morgan_fingerprints.csv')
df_MOfing = df_MOfing.drop_duplicates()
df_Rfing= pd.read_csv('/datasetNOEC/RDKitFP_fingerprints.csv')
df_Rfing = df_Rfing.drop_duplicates()
df_Tfing= pd.read_csv('/datasetNOEC/Torsion_fingerprints.csv')
df_Tfing = df_Tfing.drop_duplicates()
```

```python
# Merge all dataframes sequentially on the 'SMILES' column
merged_df = df_phys.merge(df_des, on='SMILES', how='left') \
                   .merge(df_Afing, on='SMILES', how='left') \
                   .merge(df_Mfing, on='SMILES', how='left') \
                   .merge(df_MOfing, on='SMILES', how='left') \
                   .merge(df_Rfing, on='SMILES', how='left') \
                   .merge(df_Tfing, on='SMILES', how='left')

merged_df = merged_df.drop_duplicates()
merged_df
```

```python
merged_df.to_csv('NOEC_allpropdesfing.csv', index=False)
```

# Merging with the pNOEC file

```python
import pandas as pd
df=pd.read_csv('NOEC_allpropdesfing.csv')
df2 = pd.read_csv('NOEC_meansd.csv')
# Rename the column in df2 to match df for merging
df2 = df2.rename(columns={'MolecularWeight': 'Molecular Weight'})
df2
```

```python
df['Molecular Weight'] = df['Molecular Weight'].astype(str)
df2['Molecular Weight'] = df2['Molecular Weight'].astype(str)

# Remove duplicates in 'Molecular Weight' column in df to prevent duplicate rows in the merge
df_unique = df.drop_duplicates(subset=['Molecular Weight'])

# Merge the DataFrames on the 'Molecular Weight' column, keeping only rows from df2
new_df = pd.merge(df2, df_unique, on='Molecular Weight', how='left')
new_df = new_df.dropna(subset=['CID'])
new_df
```

```python
new_df.to_csv('finalNOEC.csv', index=False)
```
