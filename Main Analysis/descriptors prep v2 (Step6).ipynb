```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import VarianceThreshold
```

```python
df_desc=pd.read_excel('noecafterknime2.xlsx')
df_desc.head()
df_desc
```

```python
df_desc_numeric = df_desc.select_dtypes(include=['number'])
df_desc_max = df_desc_numeric.max()
df_desc_min = df_desc_numeric.min()
df_norm = (df_desc_numeric - df_desc_min) / (df_desc_max - df_desc_min)
print(df_norm.head())
```

```python
#remove NA after normalisation if it appeears if min and max value are same for all molecules
df_norm =df_norm.dropna(axis=1)
df_norm
```

```python
df_norm.to_csv('Noecdescriptorsready.csv', index=False)
```
