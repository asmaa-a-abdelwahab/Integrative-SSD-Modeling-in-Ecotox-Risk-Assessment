```python
import pandas as pd
df=pd.read_csv ('Noecreadymodel.csv')
df
```

```python
col_name=df.columns[:-2]
x=df[col_name] #only descriptors
x
```

```python
y=df['pNOEC_Algae'] #endpoint
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
LR= LinearRegression()
sfs=SequentialFeatureSelector(LR, n_features_to_select=20)
sfs.fit(x, y)
```

```python
#to know the selected features
x.columns[sfs.get_support()]
sfs.get_support()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=11)
```

```python
x_train[x.columns[sfs.get_support()]] #select all the features
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.model_selection import train_test_split

# Sample data preparation (assuming x, y are defined)
# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

train_score = []
test_score = []

# Start feature selection from 2 to 14 features
for i in range(2, 20):  # Adjusting to include up to 14 features
    LR = LinearRegression()
    sfs = SequentialFeatureSelector(LR, n_features_to_select=i)
    sfs.fit(x_train, y_train)  # Make sure to fit using training data

    selected_features = x_train.columns[sfs.get_support()]
    print(f'Selected features for {i} features: {list(selected_features)}')
    
    x_train_select = x_train[selected_features]
    x_test_select = x_test[selected_features]

    LR.fit(x_train_select, y_train)
    train_accuracy = LR.score(x_train_select, y_train)
    test_accuracy = LR.score(x_test_select, y_test)
    
    train_score.append(train_accuracy)
    test_score.append(test_accuracy)
    
    print(f'{i} features - Train Score: {train_accuracy:.4f}, Test Score: {test_accuracy:.4f}')

# Optionally, visualize the scores
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(range(2, 20), train_score, label='Train Score')
plt.plot(range(2, 20), test_score, label='Test Score')
plt.xlabel('Number of Features')
plt.ylabel('R² Score')
plt.title('Model Performance with Varying Number of Features')
plt.legend()
plt.grid(True)
plt.show()
```
